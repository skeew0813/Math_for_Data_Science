# Working with Matrices

## 1.

```python
import numpy as np

# Define the matrices
A = np.array([[1, -2, 3, 7],
              [2, 1, 1, 4],
              [-3, 2, -2, 10]])

B = np.array([[0, 1, -3, -2],
              [10, -1, 2, -3],
              [5, 1, -1, 4]])

C = np.array([[4, 0, -2, 3],
              [3, 6, 9, 7],
              [2, 2, 5, 1],
              [9, 4, 6, -2]])
```

## 2.

```python
print("Dimensions of A:", A.shape)
print("Dimensions of B:", B.shape)
print("Dimensions of C:", C.shape)
```

## 3.

### (a)

```python
try:
    A_plus_B = A + B
    print("\nA + B:\n", A_plus_B)
except ValueError as e:
    print("\nA + B is not defined:", e)
```

### (b)

```python
try:
    AB = np.dot(A, B)
    print("\nAB:\n", AB)
except ValueError as e:
    print("\nAB is not defined:", e)
```

### (c)

```python
try:
    AC = np.dot(A, C)
    print("\nAC:\n", AC)
except ValueError as e:
    print("\nAC is not defined:", e)
```

### (d)

```python
C_T = C.T
print("\nTranspose of C (C^T):\n", C_T)
```

## 4.

```python
# Illustrate (A + B)^T = A^T + B^T
left_transpose = (A + B).T
right_transpose = A.T + B.T

print("\n(A + B)^T:\n", left_transpose)
print("A^T + B^T:\n", right_transpose)
```

## 5.

```python
# Illustrate (AC)^T = C^T A^T
left_ac_transpose = (np.dot(A, C)).T
right_ac_transpose = np.dot(C.T, A.T)

print("\n(AC)^T:\n", left_ac_transpose)
print("C^T A^T:\n", right_ac_transpose)
```

## 6.

```python
try:
    C_inv = np.linalg.inv(C)
    print("\nC^-1 (Inverse of C):\n", C_inv)
except np.linalg.LinAlgError as e:
    print("\nC^-1 is not defined:", e)
```
