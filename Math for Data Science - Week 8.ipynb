# 1. Solving a System of Linear Equations

```python
import numpy as np

# Define matrix A and vector b
A = np.array([[27, -10, 4, -29],
              [-16, 5, -2, 18],
              [-17, 4, -2, 20],
              [-7, 2, -1, 8]])

b = np.array([1, -1, 0, 1])

# Solve for x using numpy's linear algebra solver
x = np.linalg.solve(A, b)

# Print the solution
print(x)
```

# 2. Visualizing a Linear Transformation

## (a)

```python
# Define matrix A
A = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2],
              [np.sqrt(2)/2, np.sqrt(2)/2]])

# Define vector x
x = np.array([2, -2])

# Compute T(x) = A @ x
Tx = A @ x

# Print the result
print("T(x) =", Tx)
```

## (b)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
xx, yy = np.meshgrid(x, y)

# Define the transformation matrix A
A = np.array([
    [np.sqrt(2)/2, -np.sqrt(2)/2],
    [np.sqrt(2)/2, np.sqrt(2)/2]
])

# Stack the grid points into a 2-row matrix
xy = np.vstack([xx.flatten(), yy.flatten()])

# Apply the transformation
trans = A @ xy

# Reshape transformed points back into grid shape
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

# Plot original and transformed grids
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Original grid
axes[0].scatter(xx, yy, s=10, c=xx + yy)
axes[0].set_title("Original Grid")
axes[0].set_xlim(-10, 10)
axes[0].set_ylim(-10, 10)
axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].axvline(0, color='black', linewidth=0.5)

# Transformed grid
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx + yy)
axes[1].set_title("Transformed Grid")
axes[1].set_xlim(-10, 10)
axes[1].set_ylim(-10, 10)
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].axvline(0, color='black', linewidth=0.5)

plt.show()
```

## (c)

From the visualization in part (b), the transformation matrix 'A' appears to rotate vectors by 45° counterclockwise while preserving their lengths.
